### To fill gender with 'U' based on the distribution

```python
# calculate the proportion of 'M' and 'F' genders in the dataset excluding 'U'
prop = df_2[df_2['gender'] != 'U']['gender'].value_counts(normalize=True)

# fill the 'U' values with either 'M' or 'F' based on the proportion calculated above
df_2.loc[df_2['gender'] == 'U', 'gender'] = np.random.choice(prop.index, size=len(df_2[df_2['gender'] == 'U']), p=prop.values)

# print the updated dataset
df_2.head()
```
